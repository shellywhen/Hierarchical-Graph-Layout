```python
import json
with open ('data/DBLP.json','r')as f:
    d=json.load(f)
newNode=[]
newLink=[]
for i in d[0]['links']:
    var={}
    var['source']=i['source']
    var['target']=i['target']
    var['index']=i['index']
    newLink.append(var)
for i in d[0]['nodes']:
    var={}
    var['id']=i['id']
    var['degree']=i['degree']
    var['index']=i['index']
    newNode.append(var)
```

```python
data=[{'node':newNode,'link':newLink}]
data
```

```python
with open ('dblp.json','w')as f:
    json.dump(data,f)
```
